```python
%matplotlib inline
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten, Reshape, Lambda
import keras
import tensorflow as tf

model =  Sequential()
SEED = 1
```

```python
model.add(Dense(units=20, input_dim=10, kernel_initializer=keras.initializers.glorot_uniform(seed=SEED)))
model.add(Dropout(rate=0.2, seed = SEED))
model.add(Activation('relu'))
model.add(Reshape((4,5)))
model.add(Dense(units=10,kernel_initializer=keras.initializers.glorot_uniform(seed=SEED)))
model.add(Activation('relu'))
model.add(Flatten())
model.add(Lambda(lambda x: tf.keras.backend.sin(x)))
model.add(Dense(units=1, input_dim=40, kernel_initializer=keras.initializers.glorot_uniform(seed=SEED)))
model.add(Activation('sigmoid'))
```

```python
from keras.utils import plot_model
plot_model(model, to_file='model1.png',show_shapes=True,show_layer_names=True)
```
